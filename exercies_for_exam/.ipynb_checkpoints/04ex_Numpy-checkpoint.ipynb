2\. Find the outer product of the following two vecotrs


7\. Prime numbers sieve: compute the prime numbers in the 0-N (N=99 to start with) range with a sieve (mask).
  * Constract a shape (100,) boolean array, the mask
  * Identify the multiples of each number starting from 2 and set accordingly the corresponding mask element
  * Apply the mask to obtain an array of ordered prime numbers
  * Check the performances (timeit); how does it scale with N?
  * Implement the optimization suggested in the [sieve of Eratosthenes](https://en.wikipedia.org/wiki/Sieve_of_Eratosthenes)

````python
import time 
def naive_primes(N):
    mask = [True for i in range(N)]
    numbers = np.arange(N)
    for i in range(2,N):
        count = 2
        while i*count < N:
            mask[i*count]=False
            count+=1
    return numbers[mask][2:]

def er_primes(N):
    mask = [True for i in range(N)]
    numbers = np.arange(N)
    for i in range(2,int(np.ceil(np.sqrt(N)))):
        count = 0
        if mask[i]:
            while i**2+i*count<N:
                mask[i**2+i*count]=False
                count+=1
    return numbers[mask][2:]

size = np.arange(0,10000,1000)
time_naive = []
time_er = []

for s in size:
    
    time_in = time.time()
    [naive_primes(s) for i in range(100)]
    time_naive.append((time.time()-time_in)/100)

    time_in = time.time()
    [er_primes(s) for i in range(100)]
    time_er.append((time.time()-time_in)/100)

fig, ax = plt.subplots(1,1)
ax.plot(size,time_naive,label="naive")
ax.plot(size,time_er,label="era")
ax.legend()
````

8\. Diffusion using random walk

Consider a simple random walk process: at each step in time, a walker jumps right or left (+1 or -1) with equal probability. The goal is to find the typical distance from the origin of a random walker after a given amount of time. 
To do that, let's simulate many walkers and create a 2D array with each walker as a raw and the actual time evolution as columns

  * Take 1000 walkers and let them walk for 200 steps
  * Use randint to create a 2D array of size walkers x steps with values -1 or 1
  * Build the actual walking distances for each walker (i.e. another 2D array "summing on each raw")
  * Take the square of that 2D array (elementwise)
  * Compute the mean of the squared distances at each step (i.e. the mean along the columns)
  * Plot the average distances (sqrt(distance\*\*2)) as a function of time (step)
  
Did you get what you expected?

````python
N=1000
n_steps=200

steps = np.random.randint(0,2,size=(N,n_steps))*2-1
print("steps\n",steps)
distance = steps[:,0].reshape(N,1) # first column
for i in range(2,n_steps+1):
    distance = np.hstack([distance,np.apply_along_axis(np.sum,1,steps[:,0:i]).reshape(N,1)])
print("distance\n",distance)
distance_sqr = distance**2
print("distanceÂ²:\n",distance_sqr)
dist_mean = np.mean(distance_sqr,axis=0)
print("mean distance:", dist_mean)

fig, axis = plt.subplots(1,1)
axis.plot(range(n_steps),np.sqrt(dist_mean))
````

````python
data = np.loadtxt('populations.txt')
years=data.T[0]
hares=data.T[1]
lynxes=data.T[2]
carrots=data.T[3]

print(data)

sets = [hares, lynxes, carrots]
labels= ['hares', 'lynxes', 'carrots']

fig, ax = plt.subplots(1,3,figsize=(16,5))

for s,l,a in zip(sets,labels,ax):
    a.plot(years,s,label=l)
    a.set_xlabel("year")
    a.set_ylabel("counts")
    a.set_title(l)
    plt.setp(a.get_xticklabels(), rotation=30)

print("##### brief statistical analysis #######")
for s, l in zip(sets,labels):
    print(l,"mean:", round(np.mean(s),3),"std:", round(np.std(s),3))

print("correlation matrix hares lynxes:\n",np.corrcoef(hares,lynxes))
print("correlation matrix carrots lynxes:\n",np.corrcoef(carrots,lynxes))
print("correlation matrix carrots hares:\n",np.corrcoef(carrots,hares))

print("\n##### max population each year######\n")

index=np.apply_along_axis(np.argmax,1,data[:,1:])
print(index)
for i in range(len(index)):
    print("max population", data[i,1+index[i]], "of", labels[index[i]])
````

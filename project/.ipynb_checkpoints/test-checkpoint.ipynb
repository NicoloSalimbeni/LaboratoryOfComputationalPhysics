```python
############  HISTO 2D ###############

import matplotlib.pyplot as plt
import numpy as np

# Dati di esempio
x = np.random.uniform(0, 2, 100)
y = np.random.uniform(0, 3, 100)

# Definisci i limiti dei bin lungo l'asse x e y
bins_x = np.linspace(0, 2, 3)  # 2 bin lungo l'asse x
bins_y = np.linspace(0, 3, 4)  # 3 bin lungo l'asse y

# Crea una figura e un oggetto Axes
fig, ax = plt.subplots()

# Usa ax.hist2d per creare un istogramma bidimensionale
h = ax.hist2d(x, y, bins=[bins_x, bins_y], cmap='viridis')

print(h)

# Aggiungi etichette e titoli
ax.set_xlabel('Asse X')
ax.set_ylabel('Asse Y')
ax.set_title('Istogramma Bidimensionale')

# Aggiungi una barra del colore per la scala
plt.colorbar(h[3], ax=ax)

# Mostra il grafico
```
